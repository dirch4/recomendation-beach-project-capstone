# Cleaning Data

```python
# Drop columns
columns_to_drop = ['review_id', 'published_at', 'review_link', 'name', 'reviewer_id', 'reviewer_profile', 'published_at_date',
                   'response_from_owner_text', 'response_from_owner_ago', 'response_from_owner_date',
                   'total_number_of_reviews_by_reviewer', 'total_number_of_photos_by_reviewer', 'is_local_guide',
                   'review_translated_text', 'response_from_owner_translated_text', 'experience_details', 'review_photos']
df_cleaned = df_all.drop(columns=columns_to_drop, errors='ignore')

df_cleaned.info()
```

```python
# Check for NaN values in the 'rating' column
nan_count_rating = df_cleaned['rating'].isnull().sum()
print(f"\nNumber of NaN values in 'rating' column: {nan_count_rating}")
```

```python
# Check for NaN values in the 'review_text' column
nan_count_rating = df_cleaned['review_text'].isnull().sum()
print(f"\nNumber of NaN values in 'review_text' column: {nan_count_rating}")
```

```python
# Drop rows with NaN values in 'review_text' and 'rating'
df_cleaned = df_cleaned.dropna(subset=['review_text', 'rating'])

# Verify the changes
nan_count_rating = df_cleaned['rating'].isnull().sum()
print(f"\nNumber of NaN values in 'rating' column after dropping rows: {nan_count_rating}")

nan_count_review_text = df_cleaned['review_text'].isnull().sum()
print(f"\nNumber of NaN values in 'review_text' column after dropping rows: {nan_count_review_text}")
```

```python
from langdetect import detect, LangDetectException


def detect_language(text):
  try:
    return detect(text)
  except LangDetectException:
    return 'unknown'

df_cleaned['language'] = df_cleaned['review_text'].apply(detect_language)

# Now you can see the detected language for each review
print(df_cleaned[['review_text', 'language']].head())

# You can also get counts of each language
print(df_cleaned['language'].value_counts())
```

```python
from googletrans import Translator


# Initialize the translator
translator = Translator()

def translate_to_indonesian(text):
  try:
    if detect_language(text) != 'id':
      translated = translator.translate(text, dest='id')
      return translated.text
    else:
      return text  # Return original text if already in Indonesian
  except Exception as e:
    print(f"Translation error: {e}")
    return text # Return original text if translation fails
```

```python
# Apply the translation function to the 'review_text' column
df_cleaned['translated_review_text'] = df_cleaned['review_text'].apply(translate_to_indonesian)

# Now you have a new column 'translated_review_text' with Indonesian translations
print(df_cleaned[['review_text', 'translated_review_text', 'language']].head())
```

# Preprocessing

```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
!pip install sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

from wordcloud import WordCloud  # Membuat visualisasi berbentuk awan kata (word cloud) dari teks
```

```python
def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text) # remove hashtag
    text = re.sub(r'RT[\s]', '', text) # remove RT
    text = re.sub(r"http\S+", '', text) # remove link
    text = re.sub(r'[0-9]+', '', text) # remove numbers
    text = re.sub(r'[^\w\s]', '', text) # remove numbers
    text = text.replace('\n', ' ') # replace new line into space
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove all punctuations
    text = text.strip(' ') # remove characters space from both left and right text
    return text

def casefoldingText(text): # Converting text into lower case
    text = text.lower()
    return text

def tokenizingText(text): # Tokenizing
    text = word_tokenize(text)
    return text

def filteringText(text): # Remove stopwors in a text
    listStopwords = set(stopwords.words('indonesian'))
    listStopwords1 = set(stopwords.words('english'))
    listStopwords.update(listStopwords1)
    listStopwords.update(['iya','yaa','gak','nya','na','sih','ku',"di","ga","ya","gaa","loh","kah","woi","woii","woy"])
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered
    return text

def stemmingText(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    # Memecah teks
    words = text.split()
    # Menerapkan stemming
    stemmed_words = [stemmer.stem(word) for word in words]
    # Menggabungkan kata-kata yang telah distem
    stemmed_text = ' '.join(stemmed_words)
    return stemmed_text

    return stemmed_text
def toSentence(list_words): # Convert list of words into sentence
    sentence = ' '.join(word for word in list_words)
    return sentence
```
